```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('Attrition data.csv')
df
```

```python
df.info()
```

```python
from sklearn import preprocessing 

obj = list(df.select_dtypes(include=['object']).columns)
df[obj]
label_encoder = preprocessing.LabelEncoder() 
for col in obj:
    df[col] = label_encoder.fit_transform(df[col])
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["NumCompaniesWorked"].fillna( method ='ffill', inplace = True) 
df["EnvironmentSatisfaction"].fillna( method ='ffill', inplace = True)
df["JobSatisfaction"].fillna( method ='ffill', inplace = True)
df["WorkLifeBalance"].fillna( method ='ffill', inplace = True)
```

```python
df.isnull().sum()
```

```python
df=df.drop(['Over18','StandardHours','EmployeeCount'],axis=1)
df
```

```python
df=df.drop(['PercentSalaryHike','Age','YearsWithCurrManager'],axis=1)
df
```

```python
zero_counts = (df == 0).sum()
columns_with_zero = zero_counts[zero_counts > 0]
print(columns_with_zero)
```

```python
columns_to_replace = ['Attrition','BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','NumCompaniesWorked', 
   'StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','YearsAtCompany','YearsSinceLastPromotion']
df[columns_to_replace] = df[columns_to_replace].replace(0,1).fillna(method='ffill')
df
```

```python
columns_to_replace = ['Attrition','Gender','JobRole','StockOptionLevel','YearsSinceLastPromotion']
df[columns_to_replace] = df[columns_to_replace].replace(0, np.nan).fillna(method='ffill')
df
```

```python
nan_in_columns = df.isnull().any()

print("NaN values present in each column:\n", nan_in_columns)
```

```python
X = df.iloc[:,:-1].values
Y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size=0.7, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=42, max_iter=5000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print('Y_pred', y_pred)
print('Y_test', y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

accuracy=accuracy_score(y_test,y_pred)*100
print("Accuracy of the model is {:.2f}".format(accuracy))
```

```python
from sklearn.metrics import classification_report
cf1=classification_report(y_pred,y_test)

print('Classification_report:\n', cf1)
```
